```python
def uni_variate_feature_selection(model_name, train_scaled_data, train_class):
    """
    :param model_name: Name of  baseline model to use
    :param train_scaled_data: The min-max/normalized data
    :param train_class: The target value
    """
    # Need to use minMaxScaler for chi2 and SelectKBest// NO NEGATIVE VALUES
    # This reduces accuracy
    # Select top 4/ as max. accuracy obtained with 16 indexes removed, leaving 4 so take the 4
    k_best_features = SelectKBest(chi2, k=4).fit_transform(train_scaled_data, train_class)
    print(k_best_features)
    data_train, data_test, class_train, class_test = main.split_data(k_best_features, train_class)
    model = best_models(model_name)
    model.fit(data_train, class_train)
    prediction = model.predict(data_test)
    print(model_name + " Confusion matrix univariate:\n {}".
          format(metrics.confusion_matrix(class_test, prediction)))
    print(model_name + " Classification report  univariate :\n {}".
          format(metrics.classification_report(class_test, prediction)))
    features_scores = cross_val_score(model, data_train, class_train, cv=10)
    print(model_name+" Univariate Mean scores : {}".format(features_scores.mean()))
    print("End of "+model_name+" univariate feature selection with optimized hyperparameters")
```

```python
def best_models(model_name):
    if model_name == "SVM":
        model = SVC(kernel='linear', C=10, gamma=0.001)
    elif model_name == "RF":
        model = RandomForestClassifier(max_depth=55, n_estimators=1000)
    else:
        model = KNeighborsClassifier(n_neighbors=73, p=1)
    return model
```

```python
uni_variate_feature_selection("SVM", train_features_scaled, train_class)
```

```python
def optimize(model, data_train, class_train):
    """
    GridCV for the top performing models after two rounds of classification
    1. Without Feature Selection/Importance
    2. With Feature Importance
    :param model: name of model
    :param data_train:
    :param class_train:
    :return:
    """
    start_time = time.time()
    if model == "SVM":
        print("$$$$$$$$$$$$$$$$$$$$$ BEGINNING SVM Hyperparameter optimization $$$$$$$$$$$$$$$$$$$$")
        svm_param_grid = [{'kernel': ['linear', 'rbf'], 'C': [0.001, 0.01, 0.1, 1, 10], 'gamma':[0.001, 0.01, 0.1, 1]}]
        svm_clf = GridSearchCV(SVC(), svm_param_grid, n_jobs=-1, cv=10)
        svm_clf.fit(data_train, class_train)
        print("Best Parameters with SVM Linear Kernal:")
        print(svm_clf.best_params_, 'with a score of ', svm_clf.best_score_)
        print("$$$$$$$$$$$$$$$$$$$$$ ENDING SVM Hyperparameter optimization $$$$$$$$$$$$$$$$$$$$")
        # Best Parameters with SVM RBF Kernel:
        # {'C': 10, 'gamma': 0.001} with a score of  0.9181250000000001
        # Best Parameters with SVM Linear Kernal:
        # {'C': 1, 'gamma': 0.001} with a score of  0.959375
    return svm_clf.best_params_, svm_clf.best_score_

best_param, best_score = optimize("SVM", data_train, class_train)
```

```python
test = []
for k,v in best_param.items():
    test.append((k,v))
print(test)
config = ""
for item in test:
    config = item[0]+"= "+str(item[1])
print(config)
```
